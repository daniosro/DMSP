Let's add a transcripts per million sequences column to the dataframe, which is one of different types of normalization that we can apply to the data

```python
#Filter by total
df_total = df_stab8_cursonetal_2018[(df_stab8_cursonetal_2018.Enzyme == 'Total')]

#Get the total transcript number
transcript_number = df_total.Number_of_transcripts.sum()

#Initialize empty lists
transcripts=[]
ttpms = []

#Loop through rows in dataframe
for index, row in df_stab8_cursonetal_2018.iterrows():
    # Extract Sample_ID 
    Sample_ID = row.Sample_ID
    # Extract number of transcripts from total Dataframe
    transcript = df_total[df_total.Sample_ID == Sample_ID].Number_of_transcripts.iloc[0]
    # Compute Total_transcripts_per_million_sequences
    transcript_ratio = row.Number_of_transcripts * 1E6 / transcript 
    # Append to list
    ttpms.append(transcript_ratio)

# Create column of total_transcripts_per_million_sequences in dataframe
df_stab8_cursonetal_2018['Total_transcripts_per_million_sequences'] = ttpms
#Replace Alma1 for small initials
df_stab8_cursonetal_2018['Enzyme'] = df_stab8_cursonetal_2018['Enzyme'].replace(['Alma1'], 'alma1')
df_stab8_cursonetal_2018.head()
```

We will perform a different normalization of the above transcript data in order to be able to better compare them. For this, we will express them in terms of reads per kilobase per million mapped reads (RPKM), which is equivalent to:

$$
RPKM = {Number\;of\;reads\;mapped\;to\;a\;gene \times 10^3 \times 10^6 \over total\;number\;of\;mapped\;reads\;from\;the\;library \times gene\;length\;in\;base\;pairs}
$$

Here, $10^3$ normalizes for gene length and $10^6$ for sequencing depth factor.
let's create a dataframe to include the length of each gene in base paris.

```python
# Create dataframe for gene length in base pairs
gene_length = {'Gene':['alma1', 'dddD', 'dddK', 'dddL', 'dddP', 'dddW', 'dddY', 'dddQ', 'dmdA', 'dsyb'], 
               'Length':[1380, 2481, 756, 672, 1260, 465, 1203, 576, 1140,1050]}
df_glength = pd.DataFrame(gene_length)
df_glength
```

We will filter the transcripts dataframe to include only the enzymes of interest.

```python
#Filter to include only enzymes
df_enzymes = df_stab8_cursonetal_2018[(df_stab8_cursonetal_2018.Enzyme != 'Total')
                                     & (df_stab8_cursonetal_2018.Enzyme != 'dsyB')
                                     & (df_stab8_cursonetal_2018.Enzyme != 'DSYB')]
df_enzymes.head()
```

Now, we will calculate the reads per kilo base per million mapped reads and append them to a column in the transcripts dataframe.

```python
#Calculate reads per kilo base per million mapped reads
#Initialize empty list
rpkm_=[]

#Loop through rows in dataframe
for index, row in df_enzymes.iterrows():
    # Extract Enzyme name 
    Enzyme = row.Enzyme
    # Extract lengths from gene length Dataframe
    gl = df_glength[df_glength.Gene == Enzyme].Length.values[0]
    # Compute RPKM (Reads per kilo base per million mapped reads)
    rpkm = row.Total_transcripts_per_million_sequences * 1E3 / gl
    # Append to list
    rpkm_.append(rpkm)
# Create column of reads per kilo base per million mapped reads in dataframe
df_enzymes = df_enzymes.assign(
    Reads_per_Kb_per_million_mapped_reads = rpkm_
)

df_enzymes.head()
```

```python
#Define lists of unique values
enzymes = df_enzymes[('Enzyme')].unique()
depth_zones = df_enzymes[('Depth_zone')].unique()
ocean_and_sea_regions = df_enzymes[('Ocean_and_sea_regions')].unique()
marine_pelagic_biomes = df_enzymes[('Marine_pelagic_biomes _Longhurst_2007')].unique()
```

Let's visualize the transcriptomic data of Alma1, the only eukaryotic DMSP lyase. We will start by loading the data.

```python
df_stab3_vorobevetal_2020 = pd.read_csv('/Users/daniosro/git/DMSP/data/raw/genetics/stab3_vorobevetal_2020.csv')
df_stab3_vorobevetal_2020.head()
```

We will add a column of ocean and sea regions to the dataframe, to include the classification of each station according to its location.

```python
#Create ocean and sea regions column

#Initialize empty lists

stations=[]

#Loop through rows in dataframe
for index, row in df_stab3_vorobevetal_2020.iterrows():
    # Extract Station
    Station = row.Station
    # Extract number of transcripts from total Dataframe
    if 1<= Station <=30 or Station ==153:
        stations.append('Mediterranean Sea')
    elif 128<= Station <=141 or 103<= Station <=109:
        stations.append('North Pacific Ocean')
    elif 91<= Station <=102 or 110<= Station <=127:
        stations.append('South Pacific Ocean')
    elif 80<= Station <=90:
        stations.append('Southern Ocean (S of 40S)')
    elif 66<= Station <=79:
        stations.append('South Atlantic Ocean')
    elif 31<= Station <=65:
        stations.append('Indian Ocean')
    elif 142<= Station <=155:
        stations.append('North Atlantic Ocean')
    elif 158<= Station <=208:
        stations.append('Arctic Ocean')
# Create column of Ocean and sea regions
df_stab3_vorobevetal_2020['Ocean_and_sea_regions'] = stations
df_stab3_vorobevetal_2020
```

Now, we will calculate the reads per kilobase per million mapped reads (RPKM) in the eukaryotic transcriptome of the Tara Oceans expedition.

```python
df_stab3_vorobevetal_2020['Reads_per_Kb_per_million_mapped_reads'] = df_stab3_vorobevetal_2020['Transcripts'] * 1E9
df_stab3_vorobevetal_2020
```

In order to filter out the data from the Vorobev et al . (2020) study that does not correspond to the same stations of the Curson et al. (2018) study, we will add a new ID column to the previous dataframe that allows to make a match with the corresponding stations of the Curson et al. (2018) dataframe. We will start by loading a dataframe with the ID-station equivalences.

```python
df_equiv = pd.read_csv('/Users/daniosro/git/DMSP/data/raw/genetics/tara_sts_id.csv')
df_equiv.head()
```

Now, we will append a column to the transcripts dataframe with the ID corresponding to each station.

```python
#Initialize empty lists
ids=[]

#Loop through rows in dataframe
for index, row in df_stab8_cursonetal_2018.iterrows():
    # Extract id
    ide = row.Sample_ID
    # Extract ids from equivalence Dataframe
    eq_station = df_equiv[df_equiv.id == ide].station.iloc[0]
    # Append to list
    ids.append(eq_station)

#Append to dataframe
df_stab8_cursonetal_2018['Station'] = ids
df_stab8_cursonetal_2018.head()
```
